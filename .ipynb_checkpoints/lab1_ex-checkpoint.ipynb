```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("Lab1-Kafka")
    .master("local[*]")
    .config(
        "spark.jars.packages",
        "org.apache.spark:spark-sql-kafka-0-10_2.13:4.0.1,org.apache.kafka:kafka-clients:3.6.0"
    )
    .config("spark.driver.memory", "4g")
    .getOrCreate()
)

spark
```

```python
from pyspark.sql.functions import col, from_json, avg, count
from pyspark.sql.types import *
```

```python
bootstrap_servers = "10.1.1.10:30090,10.1.1.27:30091,10.1.1.203:30092"
```

Task 1: READ FROM KAFKA

```python
df_ratings_raw = (
    spark.read
    .format("kafka")
    .option("kafka.bootstrap.servers", bootstrap_servers)
    .option("subscribe", "ratings")
    .option("startingOffsets", "earliest")
    .load()
)

df_ratings_raw.printSchema()
df_ratings_raw.show(5, False)
```

```python
df_ratings_str = df_ratings_raw.selectExpr("CAST(value AS STRING) as json_str")
df_ratings_str.show(5, False)
```
